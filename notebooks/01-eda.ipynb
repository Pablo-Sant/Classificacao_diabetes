```python
data_diabetes_faixa_etaria = dataset["Diabetes"]
```

```python
data_diabetes_faixa_etaria = pd.DataFrame(data_diabetes_faixa_etaria)
```

```python
def classificar_idade(idade):
    if idade <= 30:
        return "jovem"
    elif (idade > 30) & (idade < 60):
        return "meia_idade"
    else:
        return "idoso"
```

```python
data_diabetes_faixa_etaria['faixa_etaria'] = dataset["Age"].apply(classificar_idade)
```
