##### Tratamento de valores nulos consumo de álcool:
    
    #### Ao me deparar com essa situação, pensei em 3 opções: apagar a coluna, apagar as linhas, substituir os nulos pela moda


    ##### Apagar a coluna: Não é interessante pq consumo de álcool tem forte correlação com diabetes

    ##### Apagar as linhas: Não é interessante, pois iria perder mais de 30% do dataset

    ##### Substituir os nulos pela moda: Ação mais inteligente pq preserva o tamanho do dataset e é preciso pegar a moda são variáveis categóricas

```python
dataset = dataset.drop("Unnamed: 0", axis=1)
```

```python
x_previsoras = dataset.iloc[:,0:18]
```

```python
x_previsoras.columns
```

```python
y_classe = dataset.iloc[:,18]
```

```python
y_classe
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_previsoras, y_classe, test_size=0.25, random_state=0)
```

```python
categoricas = x_train.select_dtypes(include=["object"])
indices_categoricos_train = [x_train.columns.get_loc(i) for i in categoricas]
```

```python
indices_categoricos_train
```

```python
x_train.columns
```

```python
numericas = x_train.select_dtypes(include=["int", "float"])
indices_numericos_train = [x_train.columns.get_loc(i) for i in numericas]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
```

```python
preprocessor = ColumnTransformer([
    ('onehot', OneHotEncoder(), indices_categoricos_train),
    ('scaler', StandardScaler(), indices_numericos_train)
])
```

```python
x_train = preprocessor.fit_transform(x_train)
x_test = preprocessor.transform(x_test)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
type(x_train)
```

```python
feature_names = preprocessor.get_feature_names_out()
x_train_DF = pd.DataFrame(x_train, columns=feature_names)
x_test_DF = pd.DataFrame(x_test, columns=feature_names)
```

```python
x_train_DF
```

```python
x_train_DF.to_csv('../data/processed/x_train.csv', index=False)
```

```python
x_test_DF.to_csv('../data/processed/x_test.csv', index=False)
```
