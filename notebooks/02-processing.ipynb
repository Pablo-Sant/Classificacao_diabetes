##### Tratamento de valores nulos consumo de álcool:
    
    #### Ao me deparar com essa situação, pensei em 3 opções: apagar a coluna, apagar as linhas, substituir os nulos pela moda


    ##### Apagar a coluna: Não é interessante pq consumo de álcool tem forte correlação com diabetes

    ##### Apagar as linhas: Não é interessante, pois iria perder mais de 30% do dataset

    ##### Substituir os nulos pela moda: Ação mais inteligente pq preserva o tamanho do dataset e é preciso pegar a moda são variáveis categóricas

```python
dataset = dataset.drop("Unnamed: 0", axis=1)
```

```python
dataset.columns
```

```python
x_previsoras = dataset.iloc[:,0:18]
```

```python
x_previsoras.columns
```

```python
y_classe = dataset.iloc[:,18]
```

```python
y_classe
```

```python
categoricas = x_previsoras.select_dtypes(include=["object"])
indices_categoricos = [x_previsoras.columns.get_loc(i) for i in categoricas]
```

```python
indices_categoricos
```

```python
x_previsoras.columns
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
onehotencoder = ColumnTransformer(transformers=[("onehot", OneHotEncoder(), indices_categoricos)], remainder="passthrough")
```

```python
x_previsoras = onehotencoder.fit_transform(x_previsoras)
```

```python
x_previsoras
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
standard = StandardScaler(with_mean=False)

x_previsoras = standard.fit_transform(x_previsoras)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_previsoras, y_classe, test_size=0.25, random_state=0)
```
